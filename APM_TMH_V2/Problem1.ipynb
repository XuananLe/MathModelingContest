```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("TreeCoverLossByYear.csv")
df['Hectares'] = df['Hectares'] / 1000
df['Year'] = df['Year'].astype('int')
df['Rolling_Mean'] = df['Hectares'].rolling(window=2).mean()

plt.figure(figsize=(10, 5))
plt.plot(df['Year'], df['Hectares'], marker='o')
plt.title('Vietnam: Tree cover loss by year')
plt.xlabel('Year')
plt.ylabel('Hectares (Thousand)')
plt.grid(True)
y_ticks = np.linspace(df['Hectares'].min(), df['Hectares'].max(), len(df['Hectares']))

plt.plot(df['Year'], df['Rolling_Mean'], color='red', label='Rolling Mean')

plt.xticks(df['Year'].values)
plt.yticks(y_ticks)


plt.tight_layout()
plt.show()
df['Hectares'].describe().round(2)
```
