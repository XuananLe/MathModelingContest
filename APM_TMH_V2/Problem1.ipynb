```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("TreeCoverLossByYear.csv")
df['Hectares'] = df['Hectares'] / 1000
df['Year'] = df['Year'].astype('int')
df['Rolling_Mean'] = df['Hectares'].rolling(window=2).mean()

plt.figure(figsize=(10, 5))
plt.plot(df['Year'], df['Hectares'], marker='o')
plt.title('Vietnam: Tree cover loss by year')
plt.xlabel('Year')
plt.ylabel('Hectares (Thousand)')
plt.grid(True)
y_ticks = np.linspace(df['Hectares'].min(), df['Hectares'].max(), len(df['Hectares']))

plt.plot(df['Year'], df['Rolling_Mean'], color='red', label='Rolling Mean')

plt.xticks(df['Year'].values)
plt.yticks(y_ticks)


plt.tight_layout()
plt.show()
df['Hectares'].describe().round(2)
```

### Tập dữ liệu

- **Count**: Số lượng giá trị trong tập dữ liệu là 19. Điều này nghĩa là có 19 giá trị được quan sát.

- **Std**: Độ lệch chuẩn (standard deviation) là 83.304223. Một độ lệch chuẩn cao (như trong trường hợp này) cho thấy các giá trị trong tập dữ liệu phân tán rộng lạc so với giá trị trung bình.

- **Min**: Giá trị nhỏ nhất trong tập dữ liệu là 43.008000.

- **25% (tứ phân vị thứ nhất)**: 25% các giá trị trong tập dữ liệu nhỏ hơn hoặc bằng 91.218000. Điều này cũng nghĩa là 75% các giá trị lớn hơn 91.218000.

- **50% (tứ phân vị thứ hai hoặc trung vị)**: Giá trị ở giữa tập dữ liệu (khi sắp xếp theo thứ tự tăng dần) là 137.716000. Khi tập dữ liệu có số lượng lẻ (như trong trường hợp này với 19 giá trị), trung vị chính là giá trị ở giữa.

- **75% (tứ phân vị thứ ba)**: 75% các giá trị trong tập dữ liệu nhỏ hơn hoặc bằng 217.061000. Điều này cũng nghĩa là chỉ có 25% số giá trị lớn hơn con số này.

- **Max**: Giá trị lớn nhất trong tập dữ liệu là 353.446000.

### Kết luận
Nhìn chung, tập dữ liệu có sự phân tán khá lớn giữa giá trị thấp nhất và giá trị cao nhất với độ lệch chuẩn cao. Trung vị (137.716000) thấp hơn giá trị trung bình (158.822368), điều này có thể chỉ ra rằng tập dữ liệu có một số giá trị cao đặc biệt làm tăng giá trị trung bình.


```python
x = df['Year']
y = df['Hectares']

dataFrame = pd.DataFrame(df['Rolling_Mean'])
segments = [(2002, 2016), (2016, 2020)]

plt.figure(figsize=(10, 5))
for start, end in segments:
    segment = df[(df['Year'] >= start) & (df['Year'] <= end)]
    x1 = segment['Year']
    y1 = segment['Hectares']
    m, b = np.polyfit(x1, y1, 1)
    plt.plot(x1, m * x1 + b)

plt.plot(df['Year'], df['Rolling_Mean'], color='red', label='Rolling Mean')
y_ticks = np.linspace(df['Hectares'].min(), df['Hectares'].max(), len(df['Hectares']))

plt.title('Vietnam: Tree cover loss by year')
plt.xlabel('Year')
plt.ylabel('Hectares (Thousand)')
plt.grid(True)
plt.xticks(x.values)
plt.yticks(y_ticks)
plt.legend()

plt.show()
```
